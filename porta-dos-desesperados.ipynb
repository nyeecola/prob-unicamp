Esse primeiro código simula o experimento com a estratégia de sempre trocar a porta.
O resultado é o esperado, com a taxa de vitória convergindo para 2/3.

```python
import random
import matplotlib.pyplot as plt

total_won = 0
total_played = 0
ratio_list = []

for i in range(10000):
    right = random.randint(0, 2)

    choice = random.randint(0,2)
    possible_choices = [0, 1, 2]

    del possible_choices[choice]

    to_be_removed = random.randint(0,1)

    if to_be_removed == 0:
        if possible_choices[0] == right:
            del possible_choices[1]
        else:
            del possible_choices[0]

    if to_be_removed == 1:
        if possible_choices[1] == right:
            del possible_choices[0]
        else:
            del possible_choices[1]

    choice = possible_choices[0]

    if possible_choices[0] == right:
        total_won += 1
    total_played += 1
    ratio_list.append(total_won/total_played)

print(total_won / total_played)

fig = plt.figure(figsize=(10, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.ylim([0, 1])
plt.plot(ratio_list)
plt.show()
```

Esse outro código simula o experimento com a estratégia de nunca trocar a porta.
O resultado também é o esperado, com a taxa de vitória convergindo para 1/3.

```python
total_won = 0
total_played = 0
ratio_list = []

for i in range(10000):
    right = random.randint(0, 2)

    choice = random.randint(0,2)
    possible_choices = [0, 1, 2]

    if possible_choices[choice] == right:
        total_won += 1
    total_played += 1
    ratio_list.append(total_won/total_played)

print(total_won / total_played)

fig = plt.figure(figsize=(10, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.ylim([0, 1])
plt.plot(ratio_list)
plt.show()
```
